```python
# Since we are calculating for all non BNS tourists and BNS tourists, some hotels will have repeated..
# ..themselves in already created tables before, we are collecting them and putting them as one
fy = list(pondi_ful_jan.Hotels); zy =  []
for i in range(pondi_ful_jan.shape[0]-1,-1,-1):
    j =  fy.index(pondi_ful_jan.Hotels[i])
    if j != i:
        zy.append((j, i))
print(zy)

fy_2 = list(pondi_ful_feb.Hotels); zy_2 =  []
for i in range(pondi_ful_feb.shape[0]-1,-1,-1):
    j =  fy_2.index(pondi_ful_feb.Hotels[i])
    if j != i:
        zy_2.append((j, i))
print(zy_2)

fy_3 = list(pondi_ful_mar.Hotels); zy_3 =  []
for i in range(pondi_ful_mar.shape[0]-1,-1,-1):
    j =  fy_3.index(pondi_ful_mar.Hotels[i])
    if j != i:
        zy_3.append((j, i))
print(zy_3)

# above we found the repeated hotels and their quantities, below we are adding them
for t in zy:
    pondi_ful_jan.at[t[0],'BNS'] = pondi_ful_jan.iloc[t[0]].BNS + pondi_ful_jan.iloc[t[1]].BNS
    pondi_ful_jan.at[t[0],'non BNS'] = pondi_ful_jan.iloc[t[0]]['non BNS'] + pondi_ful_jan.iloc[t[1]]['non BNS']

for t in zy_2:
    pondi_ful_feb.at[t[0],'BNS'] = pondi_ful_feb.iloc[t[0]].BNS + pondi_ful_feb.iloc[t[1]].BNS
    pondi_ful_feb.at[t[0],'non BNS'] = pondi_ful_feb.iloc[t[0]]['non BNS'] + pondi_ful_feb.iloc[t[1]]['non BNS']

for t in zy_3:
    pondi_ful_mar.at[t[0],'BNS'] = pondi_ful_mar.iloc[t[0]].BNS + pondi_ful_mar.iloc[t[1]].BNS
    pondi_ful_mar.at[t[0],'non BNS'] = pondi_ful_mar.iloc[t[0]]['non BNS'] + pondi_ful_mar.iloc[t[1]]['non BNS']

# below we are dropping the copies of repeatrd hotels  
pondi_ful_jan.drop(pondi_ful_jan.index[[t[1] for t in zy]], inplace = True)
pondi_ful_feb.drop(pondi_ful_feb.index[[t[1] for t in zy]], inplace = True)
pondi_ful_mar.drop(pondi_ful_mar.index[[t[1] for t in zy]], inplace = True)
pondi_ful_jan.reset_index(inplace = True, drop = True)
pondi_ful_feb.reset_index(inplace = True, drop = True)
pondi_ful_mar.reset_index(inplace = True, drop = True)
```
